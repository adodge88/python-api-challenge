### Analysis

- Temperatures increase as the latitude coordinates get closer to the equater, the remain higher in the southern hemisphere and drop dramatically in the northern hemisphere the further from the equater the latitude coordinates get during these winter months.
- Humidity appears to be more likely in norther regions above the 40 lat line as well as near the equater. 
- Cloudiness is not affected by latitude.

### Generate Cities

```python
#Randomly select at least 500 unique (non-repeat) cities based on latitude and longitude.
#use citipy to get a list

# Latitude values are measured relative to the equator and range from -90째 at the 
#South Pole to +90째 at the North Pole. 
#Longitude values are measured relative to the prime meridian. 
#They range from -180째 when traveling west to 180째 when traveling east. 
coord_list = []

lat = numpy.random.uniform(low=-90, high=90, size=2000)
lng = numpy.random.uniform(low=-180, high=180, size=2000)
#zip the lists into a tuple
coordinates = zip(lat,lng)
# convert so it can print
coordinates = set(coordinates)

for coordinate_pair in coordinates:
    lat, lon = coordinate_pair
    coord_list.append(citipy.nearest_city(lat, lon))
    #print (lat, lon)

#convert coordinates to city names for API calls
city_list = []

for coords in coord_list:
    city = coords.city_name
    city_list.append(city)
    
#remove duplicates from list 
unique_city_list = []

for city in city_list:
    if city not in unique_city_list:
        unique_city_list.append(city)

print(len(city_list))
print(len(unique_city_list))

# #FOR TESTING
# unique_city_list = unique_city_list[:25]
# unique_city_list
```

### API Print Log

```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL - SEE HOW THE q= is blank?
query_url = f"{url}appid={key}&units={units}&q="

final_city_list = []
lat = []
lng = []
temp = []
humidity = []
wind = []
clouds = []
API_csv = []

# Loop through the list of cities and perform a request for data on each
for city in unique_city_list:
    try:
        response = requests.get(query_url + city).json()
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        temp.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        wind.append(response['wind']['speed'])
        clouds.append(response['clouds']['all'])
        final_city_list.append(response['name'])
        API_csv.append(query_url + city)
        print(query_url + city)
    except KeyError:
        continue

#PRINT OUT API CALLS
```

```python
csv_df = pd.DataFrame(API_csv)
csv_df.head()

csv_df.to_csv(r'API_Calls.csv')
```
